```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Invistico_Airline.csv")
df.head()
```

```python
df.keys()
```

```python
df.info()
```

```python
df.describe()
```

```python
# This dataset contains 129880 rows and 23 columns
df.shape
```

```python
# Only one column has missing values. 
# The name of the column having missing values is Arrival Delay in Minutes.
# the missing value if the above column is 0.3% oh the whole data.
df.isna().sum()
```

```python
# We drop the nan vlues because it was 0.3% of the actual data
df = df.dropna()
```

```python
# categorical_data contains all the column name which has object data type
categorical_data = [feature for feature in df.columns if df[feature].dtype == "O"]
categorical_data
```

```python
# numerical_data contains all the column name which has int, float data type
numerical_data = [feature for feature in df.columns if df[feature].dtype != "O"]
numerical_data
```

```python
df.corr()
```

```python
data = df.drop('satisfaction',axis=1)
data.head()
```

```python
corr_matrix = data.corr().abs()
corr_matrix
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper
```

```python
# because there are not more columns which are highly correlated 
# we will not drop the below mention column.
to_drop = [feature for feature in upper.columns if any(upper[feature] > 0.85)]
to_drop
```

```python
label_data = df['satisfaction']
data = pd.concat([data,label_data],axis=1)
```

```python
data['satisfaction']=data['satisfaction'].map({'satisfied':0,'dissatisfied':1})
```

```python
for feature in categorical_data:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
for feature in numerical_data:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
sns.pairplot(data)
```

```python
linearly_correlated_columns = [feature for feature in upper.columns if any(upper[feature] > 0.65)]
for feature in linearly_correlated_columns:
    plt.scatter(data[feature],data['satisfaction'])
    plt.xlabel(feature)
    plt.ylabel('Satisfaction')
    plt.title(feature)
    plt.show()
```

```python
linearly_correlated_columns = [column for column in upper.columns if any(upper[column] > 0.65)]
for i, feature in enumerate(linearly_correlated_columns, 1):
    plt.figure()  # forces a new figure
    sns.regplot(data=data, x=feature, y='satisfaction',logistic=True)
    print("{} plotting {} ".format(i,feature))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le
```

```python
for feature in categorical_data:
    data[feature] = le.fit_transform(data[feature])
```

```python
data
```

```python
## Independent and Dependent features
X = data.drop('satisfaction',axis=1)
y = data['satisfaction']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(X_train)
```

```python
var_thres.get_support()
```

```python
sum(var_thres.get_support())
```

```python
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
from sklearn.feature_selection import mutual_info_classif
# determine the mutual information
mutual_info = mutual_info_classif(X_train, y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns.tolist()
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50,60],'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
## Prediction
y_pred = classifier_regressor.predict(X_test)
```

```python
## Accuracy Score
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,y_test)
print(score)
```
